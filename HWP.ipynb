```python
bad_air[(bad_air['최다풍향(16방위)']>=315 )|(bad_air['최다풍향(16방위)']<=45)]
#북풍
bad_air[(bad_air['최다풍향(16방위)']>=45 )&(bad_air['최다풍향(16방위)']<=135)]
#동풍
bad_air[(bad_air['최다풍향(16방위)']>=135 )&(bad_air['최다풍향(16방위)']<=225)]
#남풍
bad_air[(bad_air['최다풍향(16방위)']>=225 )&(bad_air['최다풍향(16방위)']<=315)]
#서풍
```

```python
wind_where=['북풍','동풍','남풍','서풍']
```

```python
bad_air_count=[len(bad_air[(bad_air['최다풍향(16방위)']>=315 )|(bad_air['최다풍향(16방위)']<=45)]),
              len(bad_air[(bad_air['최다풍향(16방위)']>=45 )&(bad_air['최다풍향(16방위)']<=135)]),
              len(bad_air[(bad_air['최다풍향(16방위)']>=135 )&(bad_air['최다풍향(16방위)']<=225)]),
              len(bad_air[(bad_air['최다풍향(16방위)']>=225 )&(bad_air['최다풍향(16방위)']<=315)])]
```

```python
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.figure(figsize=(35, 6))
plt.pie(bad_air_count, labels=wind_where, autopct='%.1f%%',wedgeprops=wedgeprops)
plt.title('공기질이 안좋을 때 바람이 부는 방향')
plt.show()
#공기의 질이 좋지 않은 날은 바림이 어디서 불어 왔는지 보기 위한 그래프
```

```python
nc_air_count=[len(nc_air[(nc_air['최다풍향(16방위)']>=315 )|(nc_air['최다풍향(16방위)']<=45)]),
              len(nc_air[(nc_air['최다풍향(16방위)']>=45 )&(nc_air['최다풍향(16방위)']<=135)]),
              len(nc_air[(nc_air['최다풍향(16방위)']>=135 )&(nc_air['최다풍향(16방위)']<=225)]),
              len(nc_air[(nc_air['최다풍향(16방위)']>=225 )&(nc_air['최다풍향(16방위)']<=315)])]
```

```python
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.figure(figsize=(35, 6))
plt.pie(nc_air_count, labels=wind_where, autopct='%.1f%%',wedgeprops=wedgeprops)
plt.title('공기질이 좋을 때 바람이 부는 방향')
plt.show()
#공기의 질이 좋은날은 바림이 어디서 불어 왔는지 보기 위한 그래프
```

```python
label=['Nice','Bad']
plt.figure(figsize=(20, 5))
p1=plt.bar(nc_air['일 미세먼지 농도(㎍/㎥)'],nc_air['평균 풍속(m/s)'], label='Bad', color='b', alpha=0.3)
p2=plt.bar(bad_air['일 미세먼지 농도(㎍/㎥)'],bad_air['평균 풍속(m/s)'], label='Nice', color='r', alpha=0.3)
plt.xlabel('일 미세먼지 농도(㎍/㎥)')
plt.ylabel('평균 풍속(m/s)')
plt.title('평균 바람 세기에 따른 미세먼지의 관계(산점도)')
plt.legend((p1,p2),label)
plt.show()
#공기의 질이 좋은날과 안좋은날을 같은 그래프에서 비교하기 위하여 합쳤으며 그래프의 투명도를 낮춰 집중되어있는곳은 진하게 보이게 하였습니다.
```
